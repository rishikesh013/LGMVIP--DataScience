# Image to Pencil Sketch with Python

```python
import cv2
import matplotlib.pyplot as plt
```

reading image and converting it into normal

```python
img = cv2.imread('dog.jpg')

img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img2)
```

# Converting into gray image

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

# Converting into a negative image

```python
inverted_image = 255 - gray_image
img2 = cv2.cvtColor(inverted_image, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

# Pencil Sketch


```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
img2 = cv2.cvtColor(pencil_sketch, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```
